# Análisis Cuantitativo GDP (PBI) Paises 1980 - 2023 Por El FMI

El presente proyecto tiene por objeto, explorar los datos recogidos por el FMI del PBI por cada pais miembro del acuerdo desde 1980 hasta 2023, se incluyen datos de proyecciones de PBI hasta 2028, pero estas son estimaciones de datos.

### 1. Importación de Librerías Requeridas

```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Carga de los Datos

```python
gdpDf = pd.read_csv("datos/world-gdp-1980_2028.csv")


# Usar melt() para convertir los años en filas y pivot() para cambiar la estructura
gdpDfMelted = gdpDf.melt(id_vars='Country', var_name='Year', value_name='Valor')
print(gdpDfMelted)
gdpTransposedDf = gdpDfMelted.pivot(index='Year', columns='Country', values='Valor')
```

```python
# Convierte el dataframe a Pandas Series por pais
colGdp = gdpTransposedDf[0:]['Colombia']
eauGdp = gdpTransposedDf[0:]['United Arab Emirates']
argGdp = gdpTransposedDf[0:]['Argentina']
```

```python
plt.figure(figsize=(25,10))

years = colGdp.index.to_list();
years = list(map(int, years))
colGdpValues = colGdp.values
eauGdpValues = eauGdp.values
argGdpValues = argGdp.values
plt.plot(years, colGdpValues, marker='o', linestyle='-', color='blue', label = 'Colombian GDP')
#plt.plot(years, eauGdpValues, marker='o', linestyle='-', color='r', label = 'UAE GDP')
plt.plot(years, argGdpValues, marker='o', linestyle='-', color='green', label = 'Argentinian GDP')
plt.title('GDP vs Year')
plt.xlabel('Year')
plt.ylabel('GDP (En Millones de Dolares)')
plt.xticks(range(min(years), max(years) + 1, 2)) #Mostras los labels en X cada 5 años
plt.legend()
plt.grid(True)
plt.show()
```
